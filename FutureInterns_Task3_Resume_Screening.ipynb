```python
import pandas as pd

resume_data = {
    'Resume': [
        'Python Machine Learning Data Science SQL',
        'Java Spring Boot MySQL',
        'Python NLP Deep Learning TensorFlow',
        'Data Analyst Excel SQL Power BI',
        'Python Machine Learning NLP'
    ]
}

df = pd.DataFrame(resume_data)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(df['Resume'])

print("Shape:", X.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

job_description = ["Python Machine Learning NLP"]

similarity = cosine_similarity(
    vectorizer.transform(job_description),
    X
)

print(similarity)
```

```python
df['Score'] = similarity[0]

ranked_resumes = df.sort_values(by='Score', ascending=False)

print(ranked_resumes)
```
